## Adjusting The Spacing Around Subplots

```python
fig, axes = plt.subplots(2, 2, sharex = True, sharey = True)

# Shrex -> is set to True ensuring that x axis limits and labels are the same across subplots
# Sharey -> is set to True, subplots will now share the same y axis ensuring the limits and labels are the same

for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins=50, color='black', alpha=0.5) # 2x2 grid

fig.subplots_adjust(wspace=0, hspace=0) # horizontal and vertical spaces between subplots
```

## Ticks, Labels, and Legends

- Tick -> This parameter is the list of xtick locations. and an optional parameter. If an empty list is passed as an argument then it will removes all xticks.

- label -> This parameter contains labels to place at the given ticks locations (optional parameter).

```python
fig, ax = plt.subplots() # let's do it this way to save some space

# Generating 1000 random values from STD
ax.plot(np.random.standard_normal(1000).cumsum()) # cumsum -> cumulative sum
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation = 30, fontsize = 8)

ax.set_xlabel('Stages')
ax.set_title('Random Crazy Plot')
plt.show()
```

## Annotations and Drawing On A Subplot

I'm definetly stealing this code! This is what I need and I never thought about it!!

I will try and incorporate something like this at work. Since I will be analyzing the competitors prices as well as my company prices, I will definetly incorporate arrowprops into my visualizations as well as zooming in on a specific time range.

```python
fig, ax = plt.subplots()

data = pd.read_csv('/content/spx.csv', index_col=0, parse_dates = True)
spx = data['SPX']

spx.plot(ax = ax, color = 'black')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='red', headwidth = 4, width = 2, headlength = 4), horizontalalignment = 'left', verticalalignment = 'top')

# Zoom in specific time range -> 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Important dates in the 2008–2009 financial crisis')
```

```python
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='red', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='blue', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='green', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)

plt.show()
```

The crosstab function in pandas is used to compute a simple cross-tabulation of two (or more) factors. Essentially, it is used to create a table that shows the frequency with which certain groups of data appear. It can be used to summarize the data in a way that provides a two-dimensional view of the relationships within it.

Source: https://www.geeksforgeeks.org/pandas-crosstab-function-in-python/

```python
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts = party_counts.reindex(index=['Thur', 'Fri', 'Sat', 'Sun'])

party_counts
```

```python
party_counts = party_counts.loc[:, 2:5]

party_pcts = party_counts.div(party_counts.sum(axis = 'columns'), axis = 'index')
party_counts
```

```python
party_pcts.plot.bar(stacked=True)
```

Point plots or scatter plots can be a useful way of examining the relationship between two one-dimensional data series.

# Facet Grids and Categorical Data

One way to visualize data with many categorical variables is to use a facet grid, which is a two-dimensional layout of plots where the data is split across the plots on each axis based on the distinct values of a certain variable. seaborn has a useful built-in function catplot that simplifies making many kinds of faceted plots split by categorical variables.

```python
sns.catplot(x='day', y='tip_pct', hue='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

Instead of grouping by "time" by different bar colors within a facet, we can also expand the facet grid by adding one row per time value as shown below:

```python
sns.catplot(x='day', y='tip_pct', row='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

```python
sns.catplot(x='tip_pct', y='day', kind='box', data=tips[tips.tip_pct < 0.5])
```
